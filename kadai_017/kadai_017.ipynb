```python
class Human:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  # メソッドを定義
  def printinfo(self):
    print(f"名前は{self.name}、年齢は{self.age}才です。")

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です。")
    else:
      print(f"{self.name}は大人ではありません。")



man_list = {"鎌田":50,"山田":18}

for name,age in man_list.items():
  man = Human(name,age)
  man.printinfo()
  man.check_adult()

```
