```python
class Human:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  # メソッドを定義
  def printinfo(self):
    print(f"名前は{self.name}、年齢は{self.age}才です。")

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です。")
    else:
      print(f"{self.name}は大人ではありません。")

man_list = []

man_list.append(Human("鎌田",50))

man_list.append(Human("山田",18))


for i in range(0,2):
  man = man_list[i]
  man.printinfo()
  man.check_adult()


```
