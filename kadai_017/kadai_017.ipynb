```python
class Human:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  # メソッドを定義
  def printinfo(self):
    print(f"名前は{self.name}、年齢は{self.age}")

  def adult(self):
    if self.age >= 20:
      print("大人です。")
    else:
      print("大人ではありません。")

man_1 = Human('鎌田',50)

man_1.printinfo()
man_1.adult()

man_2 = Human('鎌田',18)

man_2.printinfo()
man_2.adult()
```
